## Exemple1 : Distribution uniforme continue sur l’intervalle [0,1] 

### Definition :
<p> Une loi est uniforme entre une valeur a et une valeur b lorsque la densité de probabilité est toujours égale sur cet intervalle et nulle en-dehors. Graphiquement, l'aire délimitée par courbe de densité, l'axe des abscisses et les valeurs minimale et maximale prises par la variable aléatoire (v.a) forme un unique rectangle (voir page propriétés de la loi uniforme). La fonction de répartition montre un accroissement linéaire entre a et b (voir page fonctions de répartition). L’espérance est (a + b) / 2, la variance (b – a)² / 12 et le coefficient d'asymétrie est nul.</p>

### loin uniform :

```python
from random import random
def loi_uniforme(a,b):
 return (a+(b-a)*random())
```

### esperance  :

```python
import statistics
import matplotlib.pyplot as plt 
def esperance(n):
    tab=[]
    for i in range(n):
     tab.append(loi_uniforme(0,1))
    fig = plt.figure(1, figsize=(15, 10))
    plt.plot(tab)
    plt.title("Esperance de loi uniforme")
    plt.show()
    return statistics.mean(tab)
```

### Variance :

```python
import statistics
import matplotlib.pyplot as plt 
def variance(n):
    tab=[]
    for i in range(n):
     tab.append(loi_uniforme(0,1))
    fig = plt.figure(1, figsize=(15, 10))
    plt.plot(tab)
    plt.title("Variance de loi uniforme")
    plt.show()
    return statistics.variance(tab)
```

## Exemple 3 :Distribution discrète uniforme

## Definition :

La loi discrète uniforme est une loi de probabilité discrète indiquant une probabilité de se réaliser identique (équiprobabilité) à chaque valeur d’un ensemble fini de valeurs possibles. C'est-à-dire si une variable aléatoire qui peut prendre n valeurs possibles k1 , k2 , …, kn, suit une loi uniforme lorsque la probabilité de n’importe quelle valeur ki  est égale à 1/n.

## Distribution discrète uniforme :

```python
import matplotlib.pyplot as plt
def f(n,a,b):
    tab=[]
    for i in range(n):
        x=loi_uniforme(a,b)
        if x<a:
            tab.append(0)
        elif x>=b:
            tab.append(1)
        else:
            tab.append((1+abs(x-a))/(b-a+1))
    fig = plt.figure(1, figsize=(15, 10))
    plt.plot(tab)
    plt.show()
    return tab
```
